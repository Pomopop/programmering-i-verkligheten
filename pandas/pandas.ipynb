```python
import pandas as pd 
import numpy as np

df = pd.read_csv("Matteprov.csv")

#print(df.head(3)) # PRINTa de första 3
#print("\n")
#print(df.tail(2)) #printa de sista 2
#print("\n")
#print(df.dropna()) # slänger rader som saknar innehåll
#print("\n")

df = df.fillna(0) #sätter 0 där det saknas ett värde

df["Total"] = df.sum(axis=1) # räknar ut summan av en rad.
df ["betyg"] = np.where(df["Total"]>7, "Godkänt.", "Underkänt")
df
```

```python
df_G = df[df["betyg"]=="Godkänt"]
df_U = df[df["betyg"]=="Underkänt"]
df_U
```

```python
import matplotlib.pyplot as plt

betyg = df["betyg"].value_counts()
labels = df["betyg"].unique()
plt.bar(labels, betyg, width = .4, color = ["green", "orange"])
plt.title("godkända/underkända")
```
