```python
#uppgift 1
#a)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sysselsattning.csv", header=2)
df_G = df[df["sysselsättning"]=="förvärvsarbetande"]
df_U = df_G.sum(axis =1)
df_A = df_U.sum()
df_A


#b
df_C = df_G = df[df["sysselsättning"]=="ej förvärvsarbetande"]
df_I = df_C.sum(axis=1)
df_X = df_I.sum()
df_X

#c)
data = ([df_A,df_X])
myLabels = ["Förvärvarbetande", "ej förvärvarbetande"]

plt.pie(data, labels =myLabels)
plt.title("tårtdiagram")
plt.show
```
